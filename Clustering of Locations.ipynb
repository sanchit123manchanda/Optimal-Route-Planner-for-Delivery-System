```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Delhi Addresses.csv",index_col = 0)
```

```python
data
```

```python
plt.figure(figsize = (10,8))
plt.scatter(data['Longitude'],data['Latitude'],color = 'black')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Locations")
plt.show()
```

## Clustering 

```python
##Number of clusters 
n = 4
```

```python
X = data[['Latitude',"Longitude"]].values
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = n)
```

```python
model.fit(X)
```

```python
model.predict(X)
```

```python
data['Zone'] = model.predict(X)
```

```python
data
```

```python
plt.figure(figsize = (10,8))
plt.scatter(data['Longitude'],data['Latitude'],c = data['Zone'])
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Locations")

plt.legend()
```

```python
plt.figure(figsize = (10,8))
plt.scatter(data['Longitude'],data['Latitude'],c = data['Zone'],label = list(data['Zone']))
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Locations")
plt.legend(labels = list(data['Zone']))
plt.show()
```

```python
def filter_dataframe(df,zone):
    s = df.loc[df['Zone'] == zone]
    return s
```

```python
plt.figure(figsize = (10,8))
colors = ['red','blue','yellow','green']
for i in range(4):
    zone = filter_dataframe(data,i)
    plt.scatter(zone['Longitude'],zone['Latitude'],color = colors[i],label = i)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Locations")
plt.legend()
plt.show()
```
